```python
stats = pd.read_csv('cbb/cbb26.csv')
stats = stats.drop(columns=['POSTSEASON', 'CONF', 'Unnamed: 24'])
stats
```

Inserts missing stats of the team matchups

```python
matchups = pd.read_csv('../mm/mm/mm26.csv')
for index, row in matchups.iterrows():
    # location of the team based on seed
    teamHigh = stats.loc[stats['TEAM'] == row['TEAM_higher']]

    if len(teamHigh) == 1:
        # add the data of the higher seed team
        matchups.at[index, 'SEED_higher']= teamHigh['SEED']
        matchups.at[index,'G_higher'] = teamHigh['G']
        matchups.at[index,'W_higher'] = teamHigh['W']
        matchups.at[index,'ADJOE_higher'] = teamHigh['ADJOE']
        matchups.at[index,'ADJDE_higher'] = teamHigh['ADJDE']
        matchups.at[index,'BARTHAG_higher'] = teamHigh['BARTHAG']
        matchups.at[index,'EFG_O_higher'] = teamHigh['EFG_O']
        matchups.at[index,'EFG_D_higher'] = teamHigh['EFG_D']
        matchups.at[index,'TOR_higher'] = teamHigh['TOR']
        matchups.at[index,'TORD_higher'] = teamHigh['TORD']
        matchups.at[index,'ORB_higher'] = teamHigh['ORB']
        matchups.at[index,'DRB_higher'] = teamHigh['DRB']
        matchups.at[index,'FTR_higher'] = teamHigh['FTR']
        matchups.at[index,'FTRD_higher'] = teamHigh['FTRD']
        matchups.at[index,'2P_O_higher'] = teamHigh['2P_O']
        matchups.at[index,'2P_D_higher'] = teamHigh['2P_D']
        matchups.at[index,'3P_O_higher'] = teamHigh['3P_O']
        matchups.at[index,'3P_D_higher'] = teamHigh['3P_D']
        matchups.at[index,'ADJ_T_higher'] = teamHigh['ADJ_T']
        matchups.at[index,'WAB_higher'] = teamHigh['WAB']
        matchups.at[index,'Turnover_diff_higher'] = teamHigh['TO_DIFF']
    else:
        print(teamHigh)

    # add the data of the lower seed team
    teamLow = stats.loc[stats['TEAM'] == row['TEAM_lower']]
    if len(teamLow) == 1:
        matchups.at[index, 'SEED_lower']= teamLow['SEED']
        matchups.at[index,'G_lower'] = teamLow['G']
        matchups.at[index,'W_lower'] = teamLow['W']
        matchups.at[index,'ADJOE_lower'] = teamLow['ADJOE']
        matchups.at[index,'ADJDE_lower'] = teamLow['ADJDE']
        matchups.at[index,'BARTHAG_lower'] = teamLow['BARTHAG']
        matchups.at[index,'EFG_O_lower'] = teamLow['EFG_O']
        matchups.at[index,'EFG_D_lower'] = teamLow['EFG_D']
        matchups.at[index,'TOR_lower'] = teamLow['TOR']
        matchups.at[index,'TORD_lower'] = teamLow['TORD']
        matchups.at[index,'ORB_lower'] = teamLow['ORB']
        matchups.at[index,'DRB_lower'] = teamLow['DRB']
        matchups.at[index,'FTR_lower'] = teamLow['FTR']
        matchups.at[index,'FTRD_lower'] = teamLow['FTRD']
        matchups.at[index,'2P_O_lower'] = teamLow['2P_O']
        matchups.at[index,'2P_D_lower'] = teamLow['2P_D']
        matchups.at[index,'3P_O_lower'] = teamLow['3P_O']
        matchups.at[index,'3P_D_lower'] = teamLow['3P_D']
        matchups.at[index,'ADJ_T_lower'] = teamLow['ADJ_T']
        matchups.at[index,'WAB_lower'] = teamLow['WAB']
        matchups.at[index,'Turnover_diff_lower'] = teamLow['TO_DIFF']
    else:
        print(teamLow)
```
