```python
import pandas as pd
```

Get CSV of current season stats

```python
stats = pd.read_csv('cbb/cbb26.csv')
stats = stats.drop(columns=['POSTSEASON', 'CONF', 'Unnamed: 24'])
stats
```

Get the Matchups you with to predict

```python
matchups = pd.read_csv('mm/mm26.csv')
matchups
```

Inserts missing stats of the team matchups

```python
matchups = pd.read_csv('mm/mm26.csv')
for index, row in matchups.iterrows():
    # location of the team based on seed
    teamHigh = stats.loc[stats['TEAM'] == row['TEAM_higher']]
    if len(teamHigh) == 1:
        # add the data of the higher seed team
        matchups.at[index, 'SEED_higher']= teamHigh['SEED'].iloc[0]
        matchups.at[index,'G_higher'] = teamHigh['G'].iloc[0]
        matchups.at[index,'W_higher'] = teamHigh['W'].iloc[0]
        matchups.at[index,'ADJOE_higher'] = teamHigh['ADJOE'].iloc[0]
        matchups.at[index,'ADJDE_higher'] = teamHigh['ADJDE'].iloc[0]
        matchups.at[index,'BARTHAG_higher'] = teamHigh['BARTHAG'].iloc[0]
        matchups.at[index,'EFG_O_higher'] = teamHigh['EFG_O'].iloc[0]
        matchups.at[index,'EFG_D_higher'] = teamHigh['EFG_D'].iloc[0]
        matchups.at[index,'TOR_higher'] = teamHigh['TOR'].iloc[0]
        matchups.at[index,'TORD_higher'] = teamHigh['TORD'].iloc[0]
        matchups.at[index,'ORB_higher'] = teamHigh['ORB'].iloc[0]
        matchups.at[index,'DRB_higher'] = teamHigh['DRB'].iloc[0]
        matchups.at[index,'FTR_higher'] = teamHigh['FTR'].iloc[0]
        matchups.at[index,'FTRD_higher'] = teamHigh['FTRD'].iloc[0]
        matchups.at[index,'2P_O_higher'] = teamHigh['2P_O'].iloc[0]
        matchups.at[index,'2P_D_higher'] = teamHigh['2P_D'].iloc[0]
        matchups.at[index,'3P_O_higher'] = teamHigh['3P_O'].iloc[0]
        matchups.at[index,'3P_D_higher'] = teamHigh['3P_D'].iloc[0]
        matchups.at[index,'ADJ_T_higher'] = teamHigh['ADJ_T'].iloc[0]
        matchups.at[index,'WAB_higher'] = teamHigh['WAB'].iloc[0]
        matchups.at[index,'Turnover_diff_higher'] = teamHigh['TO_DIFF'].iloc[0]
    else:
        print(teamHigh, "higher", index)

    # add the data of the lower seed team
    teamLow = stats.loc[stats['TEAM'] == row['TEAM_lower']]
    if len(teamLow) == 1:
        matchups.at[index, 'SEED_lower']= teamLow['SEED'].iloc[0]
        matchups.at[index,'G_lower'] = teamLow['G'].iloc[0]
        matchups.at[index,'W_lower'] = teamLow['W'].iloc[0]
        matchups.at[index,'ADJOE_lower'] = teamLow['ADJOE'].iloc[0]
        matchups.at[index,'ADJDE_lower'] = teamLow['ADJDE'].iloc[0]
        matchups.at[index,'BARTHAG_lower'] = teamLow['BARTHAG'].iloc[0]
        matchups.at[index,'EFG_O_lower'] = teamLow['EFG_O'].iloc[0]
        matchups.at[index,'EFG_D_lower'] = teamLow['EFG_D'].iloc[0]
        matchups.at[index,'TOR_lower'] = teamLow['TOR'].iloc[0]
        matchups.at[index,'TORD_lower'] = teamLow['TORD'].iloc[0]
        matchups.at[index,'ORB_lower'] = teamLow['ORB'].iloc[0]
        matchups.at[index,'DRB_lower'] = teamLow['DRB'].iloc[0]
        matchups.at[index,'FTR_lower'] = teamLow['FTR'].iloc[0]
        matchups.at[index,'FTRD_lower'] = teamLow['FTRD'].iloc[0]
        matchups.at[index,'2P_O_lower'] = teamLow['2P_O'].iloc[0]
        matchups.at[index,'2P_D_lower'] = teamLow['2P_D'].iloc[0]
        matchups.at[index,'3P_O_lower'] = teamLow['3P_O'].iloc[0]
        matchups.at[index,'3P_D_lower'] = teamLow['3P_D'].iloc[0]
        matchups.at[index,'ADJ_T_lower'] = teamLow['ADJ_T'].iloc[0]
        matchups.at[index,'WAB_lower'] = teamLow['WAB'].iloc[0]
        matchups.at[index,'Turnover_diff_lower'] = teamLow['TO_DIFF'].iloc[0]
    else:
        print(teamLow, "Lower", index)
```

```python
matchups.to_csv('matchups26.csv', index=False)
```
